```python
def generateMPFeatures(mList):
    # path count
    PC = mList[0]
    for m in mList[1:]:
        PC = np.dot(PC, m)

    # inverse path count
    mList_ = mList
    mList_.reverse()
    
    PC_ = mList_[0]
    for m in mList_[1:]:
        PC_ = np.dot(PC_, m)
    
    # normalized path count
    NPC = np.zeros((len(PC), len(PC)))
    
    rowSums = np.sum(PC, axis = 1)
    colSums = np.sum(PC, axis = 0)
    
    for i in xrange(len(PC)):
        for j in xrange(len(PC)):
            if (rowSums[i] + colSums[j] != 0):
                NPC[i,j] = (PC[i,j] + PC_[j,i])/(rowSums[i] + colSums[j])
    
    # random walk
    RW = np.zeros((len(PC), len(PC)))
    
    for i in xrange(len(R)):
        for j in xrange(len(R)):
            if rowSums[i] != 0:
                RW[i,j] = (PC[i,j])/rowSums[i]
    
    # symmetric random walk
    SRW = np.zeros((len(PC), len(PC)))
    
    for i in xrange(len(PC)):
        for j in xrange(len(PC)):
            SRW[i,j] = PC[i,j] + PC_[j,i]
    
    return (PC, NPC, RW, SRW)
```
